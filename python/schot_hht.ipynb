```python
filename1 = './data/decays/RSA51-2014.10.02.08.56.44.571.TIQ'
#filename1 = './data/U238/RSA51/238U90+191MeVu_72uA_20141023_145852.TIQ'
dic1,_ = read_tiq(filename1, 1, 1024, 1)
x1 = dic1['data']
fs1 = dic1['fs']
t1 = np.arange(0, len(x1)/fs1, 1/fs1)
lframes1 = dic1['lframes']
center1 = dic1['center']
nframes_tot1= dic1['nframes_tot']
```

```python
filename2 = './data/decays/RSA51-2014.10.02.08.56.44.571.TIQ'
#filename2 = './data/U238/RSA52/238U90+191MeVu_72uA_20141023_145907.TIQ'
dic2,_ = read_tiq(filename2, 1, 1024, 1)
x2 = dic2['data']
fs2 = dic2['fs']
t2 = np.arange(0, len(x2)/fs2, 1/fs2)
lframes2 = dic2['lframes']
center2 = dic2['center']
nframes_tot2= dic2['nframes_tot']
```
