```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from pylab import specgram, psd
from iqtools import *
```

```python
t1, x1 = make_signal(1, 1000, l=1, noise= False)
xbar1, ins_ph1 = make_analytical(x1)
```

```python
plt.subplot(221)
plt.plot(t1, x1)
plt.subplot(222)
plt.plot(t1, ins_ph1)
plt.subplot(223)
plot_hilbert(xbar1)
```

```python
t2, x2 = make_signal(1, 1000, l=1, noise= False)
xbar2, ins_ph2 = make_analytical(x2)
```

```python
plt.subplot(221)
plt.plot(t2, x2)
plt.subplot(222)
plt.plot(t2, ins_ph2)
plt.subplot(223)
plot_hilbert(xbar2)
```

```python
plt.plot(t1, ins_ph1 - ins_ph2)
plt.grid(True)
```

```python
ph = np.angle(x1*x2.conjugate())
```

```python
plt.plot(t1, ph)
plt.grid(True)
```
