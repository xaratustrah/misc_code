# Importing Field files from MWS

#### Data structure from MWS
First three columns are x y and z resp. The rest is fields values.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
class RoQ:
    def __init__(self, filename, f0):
        self.f0 = f0
        self.roq = 0
        self.filename = filename
        self.data = np.genfromtxt(self.filename, dtype=float, skip_header=2)

    def __call__(self, x, y):
        self.datacut = np.array([row for row in self.data if row[0]==x and row[1]==y])
        return np.trapz(np.abs(self.datacut[:,5]), self.datacut[:,2]/100)**2 / 382.737e6/2/np.pi

    def plotEz(self):
        plt.plot(self.datacut[:,2], np.abs(self.datacut[:,5]))
        plt.xlabel('Z offset [cm]')
        plt.ylabel('EzRe [V/m]')
        plt.grid(True)
        plt.savefig(self.filename + '.pdf')

    def __str__(self):
        return 'The R/Q is {} ohms.'.format(self.roq)
```

```python
r = RoQ('fields.txt', 382.737e6)
```

```python
r(0,0)
```

```python
r.plotEz()
```

```python
x, y, z, E= r.data[:,0]/100, r.data[:,1]/100, r.data[:,2]/100, abs(r.data[:,5])
```

```python
x=x.reshape((21,121,121))
y=y.reshape((21,121,121))
z=z.reshape((21,121,121))
E=E.reshape((21,121,121))
```

```python
roq = np.zeros((121,121))
for i in range(121):
    for j in range(121):
        roq[i,j] = np.trapz(E[:,i,j], z[:,i,j])
roq = roq**2/r.f0/2/np.pi
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(x[0], y[0], roq)
```

```python
fig=plt.figure()
plt.pcolormesh(x[0], y[0], roq, cmap = cm.gist_heat)
plt.colorbar()
```
