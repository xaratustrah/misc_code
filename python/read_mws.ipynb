# Importing Field files from MWS

#### Data structure from MWS
First three columns are x y and z resp. The rest is fields values.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import os
```

```python
class RoQ:
    def __init__(self, filename, f0):
        self.f0 = f0
        self.roq = 0
        self.filename = filename
        self.filename_woe = os.path.splitext(self.filename)[0]
        self.data = np.genfromtxt(self.filename, dtype=float, skip_header=2)

    def __call__(self, x, y):
        self.datacut = np.array([row for row in self.data if row[0]==x and row[1]==y])
        return np.trapz(np.abs(self.datacut[:,5]), self.datacut[:,2]/100)**2 / 382.737e6/2/np.pi

    def plotEz(self):
        plt.plot(self.datacut[:,2], np.abs(self.datacut[:,5]))
        plt.xlabel('Z offset [cm]')
        plt.ylabel('EzRe [V/m]')
        plt.grid(True)
        plt.savefig(self.filename_woe + '_cut.pdf')

    def __str__(self):
        return 'The R/Q is {} ohms.'.format(self.roq)
```

```python
r = RoQ('fields.txt', 382.737e6)
```

```python
r(0,0)
```

```python
r.plotEz()
```

```python
x, y, z, E= r.data[:,0]/100, r.data[:,1]/100, r.data[:,2]/100, abs(r.data[:,5])
xmin = -30
ymin = -30
zmin = -5
xnum = int(np.abs(xmin) * 2 / 0.5 +1)
ynum = int(np.abs(ymin) * 2 / 0.5 +1)
znum = int(np.abs(zmin) * 2 / 0.5 +1)
```

```python
# reshape on z, then y, then x, x being the fastest changing in the nested loop xyz format
x=x.reshape(znum,ynum,xnum)
y=y.reshape(znum,ynum,xnum)
z=z.reshape(znum,ynum,xnum)
E=E.reshape(znum,ynum,xnum)
```

```python
# calculate integral of E along z for every fixed x and y
roq = np.zeros((xnum,ynum))
for i in range(ynum):
    for j in range(xnum):
        roq[i,j] = np.trapz(E[:,i,j], z[:,i,j])
roq = roq**2/r.f0/2/np.pi
```

```python
# set plot limits
llim=53
hlim=67
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(x[0,llim:hlim,llim:hlim], y[0,llim:hlim,llim:hlim], roq[llim:hlim,llim:hlim])
plt.xlabel('x offset [m]')
plt.ylabel('y offset [m]')
plt.savefig(r.filename_woe+'_surf.pdf')
```

```python
fig=plt.figure()
plt.pcolormesh(x[0,llim:hlim,llim:hlim], y[0,llim:hlim,llim:hlim], roq[llim:hlim,llim:hlim], cmap = cm.gist_heat)
cb= plt.colorbar()
plt.xlabel('x offset [m]')
plt.ylabel('y offset [m]')
cb.set_label('R/Q [ohm]')
plt.savefig(r.filename_woe+'_map.pdf')
```
