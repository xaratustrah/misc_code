# Importing Field files from MWS

#### Data structure from MWS
First three columns are x y and z resp. The rest is fields values.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import os
```

```python
class RoQ:
    def __init__(self, filename, f0):
        self.f0 = f0
        self.roq = 0
        self.filename = filename
        self.filename_woe = os.path.splitext(self.filename)[0]
        self.data = np.genfromtxt(self.filename, dtype=float, skip_header=2)

    def __call__(self, x, y):
        self.datacut = np.array([row for row in self.data if row[0]==x and row[1]==y])
        return np.trapz(np.abs(self.datacut[:,5]), self.datacut[:,2]/100)**2 / 382.737e6/2/np.pi

    def plotEz(self):
        plt.plot(self.datacut[:,2], np.abs(self.datacut[:,5]))
        plt.xlabel('Z offset [cm]')
        plt.ylabel('EzRe [V/m]')
        plt.grid(True)
        plt.savefig(self.filename_woe + '_cut.pdf')

    def __str__(self):
        return 'The R/Q is {} ohms.'.format(self.roq)
```

```python
#r = RoQ('fields.txt', 382.737e6)
r = RoQ('mode1e.txt', 422.838e6)
```

```python
x, y, z, E= r.data[:,0], r.data[:,1], r.data[:,2], abs(r.data[:,5])
step = 0.5
xmin = x[0]
xmax = x[-1]
xnum = int((xmax - xmin)/step + 1)
ymin = y[0]
ymax = y[-1]
ynum = int((ymax - ymin)/step + 1)
zmin = z[0]
zmax = z[-1]
znum = int((zmax - zmin)/step + 1)
```

```python
xmin, xmax, xnum, ymin, ymax, ynum, zmin, zmax, znum
```

```python
# reshape on z, then y, then x, x being the fastest changing in the nested loop xyz format
x=x.reshape(znum,ynum,xnum)
y=y.reshape(znum,ynum,xnum)
z=z.reshape(znum,ynum,xnum)
E=E.reshape(znum,ynum,xnum)
```

```python
# calculate integral of E along z for every fixed x and y
roq = np.zeros((xnum,ynum))
for i in range(xnum):
    for j in range(ynum):
        roq[i,j] = np.trapz(E[:,j,i], z[:,j,i])
roq = roq**2/r.f0/2/np.pi/1e4 # factor because of cm t meter conversion
roq = roq.T
```

```python
# set plot limits
xlo=10
xhi=30
ylo=110
yhi=130
```

```python
fig=plt.figure()
plt.pcolormesh(x[0,ylo:yhi,xlo:xhi],
               y[0,ylo:yhi,xlo:xhi], 
               roq[ylo:yhi,xlo:xhi])#, cmap = cm.gist_heat)
cb= plt.colorbar()
plt.grid(True)
plt.title('R/Q map')
plt.xlabel('x offset [cm]')
plt.ylabel('y offset [cm]')
cb.set_label('R/Q [ohm]')
plt.savefig(r.filename_woe+'_map.pdf')
plt.savefig(r.filename_woe+'_map.eps')
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_wireframe(x[0,ylo:yhi,xlo:xhi],
               y[0,ylo:yhi,xlo:xhi], 
               roq[ylo:yhi,xlo:xhi])#, cmap = cm.gist_heat)
ax.view_init(elev=10., azim=104)
plt.title('R/Q surface')
plt.xlabel('x offset [cm]')
plt.ylabel('y offset [cm]') 
plt.savefig(r.filename_woe+"_surf.eps")
plt.savefig(r.filename_woe+"_surf.pdf")
```
