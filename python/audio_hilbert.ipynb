# Hilbert transform of Audio signal

```python
import pyaudio
import numpy as np
from iqtools import *
from pylab import specgram, psd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
CHUNK = 1024
RATE = 44100
RECORD_SECONDS = 1
```

```python
p = pyaudio.PyAudio()
stream = p.open(format=pyaudio.paFloat32, channels=2, rate=RATE, input=True, frames_per_buffer=CHUNK)
for i in range(0, int(RATE / CHUNK * RECORD_SECONDS)):
    data = stream.read(CHUNK)
stream.stop_stream()
stream.close()
p.terminate()
```

```python
ar = np.fromstring(data, dtype=np.float32)
t = np.arange(0, RECORD_SECONDS, 1/ RATE)
xbar , ins_ph = make_analytical(ar)
```

```python
plt.grid(True)
plt.subplot(221)
plt.plot(ar)
plt.subplot(222)
plot_hilbert(xbar)
plt.subplot(223)
pxx, freq = psd(ar, NFFT = CHUNK, noverlap = 256, Fs = RATE)
plt.subplot(224)
pxx, freq, bins , im = specgram(ar, NFFT  = CHUNK, noverlap = 256, Fs= RATE)
```
