# Analytical signal using Hilbert transformation

```python
from scipy.signal import hilbert
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
fs = 50    # sampling frequency
f = 12.37    # frequency
t_tot = 10    #signal length
t = np.arange(0, t_tot, 1/fs)    # time vector
x = np.cos (2*np.pi*f*t/fs)    # signal
x = x + 2*x + 3*x 
```

```python
plt.plot(t,x)
```

```python
I = np.real(hilbert(x))
Q = np.imag(hilbert(x))
```

```python
plt.plot(I,Q)
```
